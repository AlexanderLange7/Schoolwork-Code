```python
import numpy as np
import pandas as pd

# read the tool.csv file as a Pandas dataframe
tool_infile = pd.read_csv("tool.csv")

# examine the data, if you wish
tool_infile
```

```python
from sklearn.model_selection import StratifiedShuffleSplit

# use StratifiedShuffleSplit to create a training set (80% of the data)
# and a testing set (20% of the data)
#
# The label for this data is the 'tool_condition' field
#
#Use the StratifiedShuffleSplit to create a training and a testing set.
split = StratifiedShuffleSplit(n_splits = 1, test_size = 0.2, random_state = 68333)
for train_index, test_index in split.split(tool_infile, tool_infile['tool_condition']):
    train_set = tool_infile.loc[train_index]
    test_set = tool_infile.loc[test_index]
```

```python
# create X and X_test, with all the features except 'tool_condition'
X_test = train_set.drop('tool_condition', axis = 1)
X=train_set.drop('tool_condition',axis=1)

# create y and y_test, with only the 'tool_condition' values
y_test = train_set['tool_condition'].copy()
y = train_set['tool_condition'].copy()
```

```python
from sklearn.preprocessing import StandardScaler

# use StandardScaler to scale X and X_test
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline

num_pipeline = Pipeline([
    ('std_scaler', StandardScaler()),
])

X_test = num_pipeline.fit_transform(X_test)
X= num_pipeline.fit_transform(X)
```

```python
import tensorflow as tf

# Use Keras to define the DNN model:
#    52-node input layer (using Flatten)
#    <hidden layer(s) are up to you; use the advice from chapter 11>
#    1-node output layer using sigmoid activation
model = tf.keras.models.Sequential([
    tf.keras.layers.Flatten(input_shape=(52,)),
    tf.keras.layers.Dense(78, activation=tf.nn.sigmoid),
    tf.keras.layers.Dense(52,activation=tf.nn.sigmoid),
    tf.keras.layers.Dense(78, activation=tf.nn.sigmoid),
    tf.keras.layers.Dense(26, activation=tf.nn.sigmoid),
    tf.keras.layers.Dense(1, activation=tf.nn.sigmoid)
])


# Compile the model into an executable form
# optimizer = method of optimizing the weights; use Nesterov or Adam
# loss = function used to evaluate error; use 'mean_squared_error'
# metrics = what measure loss is computed over
model.compile(optimizer='Adam',
             loss='mean_squared_error',
             metrics=['accuracy'])



# Train the model; an epoch is a complete training pass
# through the entire dataset. 
model.fit(X, y, epochs=20)
```

```python
# Evaluate the model on the test data / labels, and
# print the results
model.evaluate(X_test,y_test)

```
